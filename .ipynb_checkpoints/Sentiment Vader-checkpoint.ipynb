```python
import nltk
nltk.download()
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk.sentiment.util import *
f=open("demo_nov8-20.txt","r",encoding="utf-8")

# sentences.split('\n')
# sentences = ["good movie","bad movie", "better than others"]
sid = SentimentIntensityAnalyzer()
for sentence in f:
    print(sentence)
    ss = sid.polarity_scores(sentence)
    for k in sorted(ss):
        print('{0}: {1}, '.format(k, ss[k]), end='')
    print()
f.close()
```
