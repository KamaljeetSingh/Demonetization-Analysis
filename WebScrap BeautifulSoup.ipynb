```python
import bs4 as bs
import urllib.request
```

```python
sauce = urllib.request.urlopen("https://en.wikipedia.org/wiki/Hardy_Sandhu").read()
```

```python
soup = bs.BeautifulSoup(sauce,"lxml")
# this is the source code
# print(soup) 
# this is the title tag name
print(soup.title.string)
```

```python
for links in soup.find_all('a'):
    print(links.text)
```

```python
# get links name
for links in soup.find_all('a'):
    print(links.string)
```

```python
# get href links
for links in soup.find_all('a'):
    print(links.get('href'))
```
